# ZCB and Forwards (Continous Case)

```python
import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Interest Rate from Discount Rate

The case when r = constant (flat yield rate or interest rate)

$ P(t,T) = e^{-r(T-t)} \\$

$ ln(P(t,T)) = -r(T-t) $

which gives

$ \large \boxed{r = \frac{-1}{(T-t)} ln(P(t,T))} $

If the interest rate is not constant, then this equation represents the average interest rate over this period of time.

### Plot ZCB for Flat Interest Rate

```python
def ZCB(T, r):
  return np.exp(-T*r)

r = 0.1

t = 0

T = 10

t = np.linspace(t, T, 100)

P = ZCB(t, r)
```

```python
# Let's plot

plt.plot(t, P)

plt.title("Zero Coupon Price (Discount Factor)")

plt.xlabel("Years (t)")

plt.ylabel("ZCB Price (P)")

plt.xlim(0, T)
plt.ylim(0, 1.0)

plt.show()
```

```python
# now for the flat interest rate

r = 0.1

t = 0

T = 10

time = np.linspace(t, T, 100)

r_f = r*np.ones(100)

plt.plot(time, r_f)

plt.title("Flat (Constant) Interest Rate")

plt.xlabel("Years (t)")

plt.ylabel("Yield rate")

plt.show()
```

### Forward Rates (when constant/flat interest rate)

Since the interest rate is constant (flat), the forward rate will be also constant and equal to the interest rate.

Proof:
The forward rate between two periods $ t_1 $ and $ t_2 $ is calculated as follows

$ P(t, t_1) = e^{r(t_1-t)} $

$ P(t, t_2) = e^{r(t_2-t)} $

$ P(t_1, t_2) = e^{r_f(t_2-t_1)} $  (forward discount rate)

Using the no-arbitrage principle

$ P(t, t_2) = P(t, t_1) P(t_1, t_2)$

$ e^{-r(t_2 - t)} = e^{-r(t_1 - t)}e^{-r_f(t_2 - t_1)} $

$ -r(t_2 - t) = -r(t_1 - t) -r_f(t_2 - t_1) $

$ -rt_2 + rt = -rt_1 + rt -r_f(t_2 - t_1) $

$ r_f(t_2 - t_1) = rt_2 - \cancel{rt} - rt_1 + \cancel{rt} \implies r_f(t_2 - t_1) = r(t_2 - t_1)$

$ \large \boxed{\therefore r_f = r} $

### Plot Flat Forward Curve

```python
r = 0.1

t = 0

T = 10

time = np.linspace(t, T, 100)

r_f = r*np.ones(100)

plt.plot(time, r_f)

plt.title("Forward rate when interest rates are flat")

plt.xlabel("Years (t)")

plt.ylabel("Forward rate")

plt.show()
```

## Forward Rate (general formula)

The forward rate between two periods $ t_1 $ and $ t_2 $ using the time-varying interest rate $ r(t,T) $ from time t to time T

$ P(t, t_1) = e^{-r(t,t_1)(t_1-t)} $

$ P(t, t_2) = e^{-r(t,t_2)(t_2-t)} $

$ P(t_1, t_2) = e^{-r(t_1,t_2)(t_2-t_1)} $         

$ P(t_1, t_2) $ is the forward discount rate, $ r(t_1,t_2) $ is the forward rate  

Using the no-arbitrage principle

$ P(t, t_2) = P(t, t_1) P(t_1, t_2)$

$ e^{-r(t,t_2)(t_2 - t)} = e^{-r(t,t_1)(t_1 - t)}e^{-r(t_1,t_2)(t_2 - t_1)} $

$ -r(t,t_2)(t_2 - t) = -r(t,t_1)(t_1 - t) -r(t_1,t_2)(t_2 - t_1) $

$ -r(t,t_2)t_2 + r(t,t_2)t = -r(t,t_1)t_1 + r(t,t_1)t - r(t_1,t_2)(t_2 - t_1) $

$ r(t,t_2)t_2 - r(t,t_2)t = r(t,t_1)t_1 - r(t,t_1)t + r(t_1,t_2)(t_2 - t_1) $

$ r(t_1,t_2)(t_2 - t_1) = r(t,t_2)t_2 - r(t,t_2)t - r(t,t_1)t_1 + r(t,t_1)t $

$ r(t_1,t_2)(t_2 - t_1) = r(t,t_2)(t_2 - t) - r(t,t_1)(t_1 - t) $

$ \boxed{\large r(t_1,t_2) = \frac{r(t,t_2)(t_2 - t) - r(t,t_1)(t_1 - t)}{(t_2 - t_1)}} $

Another format, since

$ P(t, t_1) = e^{-r(t,t_1)(t_1-t)} \implies r(t,t_1) = \frac{-ln(P(t, t_1))}{(t_1-t)} $

$ P(t, t_2) = e^{-r(t,t_2)(t_2-t)} \implies r(t,t_2) = \frac{-ln(P(t, t_2))}{(t_2-t)} $

Therefore 

$ \large r(t_1,t_2) = \frac{r(t,t_2)(t_2 - t) - r(t,t_1)(t_1 - t)}{(t_2 - t_1)} $

becomes

$ r(t_1,t_2) = \frac{\frac{-ln(P(t, t_2))}{(t_2-t)}(t_2 - t) + \frac{ln(P(t, t_1))}{(t_1-t)}(t_1 - t)}{(t_2 - t_1)} $

$ \boxed{ \large r(t_1,t_2) = \frac{1}{(t_2 - t_1)} ln \biggl( \frac{P(t, t_1)}{P(t, t_2)}\biggr)}  $


### Plot Forward Rate Curve

Let's use the first formula to calculate the forward rates, for a daily period, given daily interest rates

```python
time = np.linspace(t, T, 365)

a = 0.05
r = a * time

# initial time
t = 0

# final time
T = 1


P = ZCB(time, r)

Forwards = []

# calculate forward rate using the daily ZCB prices
for i in range(1, 365):
  Forwards.append((1/(time[i]-time[i-1]))*np.log(P[i-1]/P[i]))

plt.plot(time[1:365], Forwards)

plt.title("Forward Rate Curve")

plt.xlabel("Years (t)")

plt.ylabel("Forward Rate")

plt.show()
```

## Short Rate (Instantaneous Forward Rate)

The short rate, also known as the instantaneous forward rate, measures the forward rate at a point in time.

$ P(t,T) = e^{-\int_t^T f(t,x) dx} $

Take the log of both sides

$ ln(P(t,T)) = -\int_t^T f(t,x) dx $

differentiate both sides

$ \large \boxed{f(t,T) = -\frac{\partial{lnP(t,T)}}{\partial{T}}} $

## Summary

The three main equations needed are 

$ \large \boxed{P(t,T) = e^{-\int_t^T f(t,x) dx}} $

from this, we can obtain the interest rate (or yield of the ZCB)

$ \large \boxed{r = \frac{-1}{(T-t)} ln(P(t,T))} $ 
or
$ \large \boxed{r = \frac{-1}{(T-t)} \int_t^T f(t,x) dx} $

and the short rate (instantaneous forward rate)

$ \large \boxed{f(t,T) = -\frac{\partial{lnP(t,T)}}{\partial{T}}} $
